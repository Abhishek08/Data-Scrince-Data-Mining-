<b>  Categorical Variables: A categorical variable can have two or more than two values. They contains a fixed set of text values which needs to be converted into numerical values so later such dataset is used in machine learning models. They are divided into two types i.e. Nominal variable or Ordinal variable. </b>

* <b>Nominal Attributes </b> contains value which does not have an order. For example :  Fuel( Gas, Oil, wood pellet) or values like Yes/No.

* <b> Binary Attributes:</b>  Binary data has only 2 values/states. For Example yes or no, affected or unaffected, true or false. 





* <b> Ordinal Attributes</b> have an order. For example: Age groups (0–20, 21–40, 41–70) or values like (low, medium, high) 






<b> Why Encoding </b>

Most of the Machine learning models understands numerical data. To make these machine learning algorithms work for insights, the data should be in the models readable form.

Deep learning techniques such as the Artificial Neural network expect data to be numerical. Thus, categorical data must be encoded to numbers before we can use it to fit and evaluate a model.

Few ML algorithms such as Tree-based (Decision Tree, Random Forest ) do a better job in handling categorical variables. The best practice in any data science project is to transform categorical data to a numeric value.

Types of Encoding techniques for Categorical Data 

* <b>Label Encoding</b>

* <b>One-hot Encoding</b>

* <b>Ordinal Encoding</b>


 

* <b>Label Encoding</b>

Label Encoding is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering. 


```python
import pandas as pd
```

```python
"""Label Encoding using Binary Attributes"""


df = {'Gender': ['Male','Female','Male','Female'],
    'Professional': ['Doctor','Engineer','Engineer','Teacher'],
       'Salary': [44000,75000,65000,90000]
    }
df = pd.DataFrame(df)
```

```python
df
```

```python
#df['code'] = pd.factorize(df['Gender'])[0]
# converting type of columns to 'category'
df['Gender'] = df['Gender'].astype('category')
df.dtypes
# converting type of columns to 'Codes Numerical form'
df['Gender'] = df['Gender'].cat.codes
```

```python
df
```

<b> One potential issue with label encoding is that most of the time, there is no relationship of any kind between categories, while label encoding introduces a relationship. </b>

<b> Label Encodin with Nominal Attribute </b>


```python

df = {'State': ['Maharashtra','Kerla','Karnataka','Tamil Nadu'],
       'Cases': [41327,18123,34047,23975]
    }
df = pd.DataFrame(df)
df
```

```python
df.State = df.State.astype('category')
```

```python
df.dtypes
```

```python
df.State = df.State.cat.codes
```

<b> One Hot Encoding </b>

In this approach, for each category of a feature, we create a new column (sometimes called a dummy variable) with binary encoding (0 or 1) to denote whether a particular row belongs to this category. 



* One-Hot Encoding is the process of creating dummy variables.


<b> Dummy Variable trap</b>
* Dummy Variable Trap is a scenario in which variables are highly correlated to each other.
* for a categorical feature with K categories, we need only K-1 indicator variables. In our example “Color_Black” was removed


```python

df = {'Fruit': ['Apple','Graps','Banana','Kiwi'],
       'Color': ['Red','Green','Yellow','Green']
    }
df = pd.DataFrame(df)
df
```

```python
y = pd.get_dummies(df.Color, prefix='Color', drop_first=True)
```

```python
y
```

```python
df1 = pd.concat([df,y],axis=1)
df1.drop('Color', axis=1, inplace=True)
df1
```
